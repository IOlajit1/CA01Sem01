```python
def group_edu(Education_level):
    if 'No formal education' in Education_level:
        return 'No formal Education'
    elif 'Primary' in Education_level:
        return 'Primary Education'
    elif any(keyword in Education_level for keyword in ['Lower secondary', 'Upper secondary', 'Technical/vocational']):
        return 'Secondary Education'
    else:
        return 'Tertiary Education'
```

As seen above, there would be an issue with the classification as the ages are reported as strings but also the ages of 'Under 1 year' and '99 years and over' will have to be modified to fit into the age groups of 0-14 and 65+ respectively. It is also good to see all the unique values at a glance to know the best way to manipulate the string

To help with the classification, the regular expression,'re' module will be imported to be able to search for and match the strings that need to be classified. At the end of the classification, a new column, Age_group, will be created

```python
import re
import warnings
warnings.filterwarnings("ignore")
```

From above, 'warnings' is imported as a result of successive code that might produce warnings as recommendations from python but the code still functional

```python
#To match the string values to the predetermined age classification 
#Note: an additional reference to '1 year' is added as it is an exception to the 'x years' condition
age_mapping = {
    '0-14 years': [str(i) + ' years' for i in range(1, 15)] + ['Under 1 year']+ ['1 year'],
    '15-24 years': [str(i) + ' years' for i in range(15, 25)],
    '25-44 years': [str(i) + ' years' for i in range(25, 45)],
    '45-64 years': [str(i) + ' years' for i in range(45, 65)],
    '65+ years':[str(i) + ' years' for i in range(65, 100)]+ ['99 years and over']
              }

# To extract the numeric part of the 'Age' column and convert it to integers
population_sexes_ages['Age_int'] = population_sexes_ages['Age'].apply(lambda x: int(re.search(r'\d+', x).group())
                                                                      if re.search(r'\d+', x) else x)

# To create new column 'Age_group' post-mapping
population_sexes_ages['Age_group'] = population_sexes_ages['Age'].apply(lambda x: next((group for group, ages in age_mapping.items() if x in ages), None))
    
population_sexes_ages.head()
```

```python
agg_df = population_sexes_ages.groupby(['Year', 'Age_group'])['Population'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid", rc={"grid.linewidth": 0.4})
sns.lineplot(data=agg_df, x='Year', y='Population', hue='Age_group',ci=None)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Irish Population Growth by Age Group from 1926-2023')

plt.legend(title='Age Group', loc="upper right", bbox_to_anchor=(1.2, 1))

plt.xlim(1926, 2026)
plt.xticks(range(1926, 2026,10), rotation=45)
plt.ylim(500000,2000000)
plt.yticks(range(200000,2000000,500000))

def format_mix (x,pos):
    if x >= 1e6:  # Display 'M' for millions
        return f'{x/1e6:.1f}M'
    elif x >= 1e3:  # Display 'K' for thousands
        return f'{x/1e3:.0f}K'
    else:
        return str(int(x))
    
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(format_mix))

plt.show()
```

Approximating Binomial to Normal Distribution

```python
from scipy.stats import norm

k = 6 
n = 10
p = 0.0794  

# Calculate mean and standard deviation for the normal approximation
mean = n * p
std_dev = (n * p * (1 - p))**0.5

result = norm.sf(k - 0.5, loc=mean, scale=std_dev)

print(f"Probability is: {result:.8f} or {result:.6%}")
```

When a normal distribution is tried on the same problem, it gives a different number, this is because normal distributions are often best suited for larger datasets and it is assumed that the dataset used is not large enough to distribute erfect normality
In the code above 'k-0.5' is implemented as when aproximating from a discrete like binomial to continuous like normal, the probability being a summation of all relevant probabilities under the curve is never a specific single point so this way, the code utilises 0.5 as its interval boundary (k+/-5)

A function below will be written to speed up and efficiently assess a number of models. Some key notes for the function:
* It will take 3 arguments, the target and independent variables and the type of regressor
* Based on the objective of the prediction, the split will be taking the first 80% as training and the last 20% as tests. Gholamy,Kreinovich and Kosheleva (2018) site 70/30 and 80/20 as the most common splits with a preference for the 80/20. So as a result, all models will use the 80/20 split


```python

def plot_regression(X, y, regressor):
    
    split_index = int(0.8 * len(X))

    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]

    regressor.fit(X_train, y_train)

    y_train_pred = regressor.predict(X_train)
    y_test_pred = regressor.predict(X_test)

    train_accuracy = regressor.score(X_train, y_train)
    test_accuracy = regressor.score(X_test, y_test)

    plt.figure(figsize=(10, 6))
    sns.set(style="whitegrid", rc={"grid.linewidth": 0.4})
    plt.scatter(X_train, y_train, label='Training Data', color='blue')
    plt.scatter(X_test, y_test, label='Test Data', color='green')
    plt.plot(X_train, y_train_pred, label='Training Prediction', color='red', linestyle='--')
    plt.plot(X_test, y_test_pred, label='Test Prediction', color='purple', linestyle='--')
    
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title(f'Population prediction using {regressor}')
    plt.legend(loc='upper right',bbox_to_anchor=(1.25,1.02))

    print(f"Training Accuracy (R^2): {train_accuracy:.2f}")
    print(f"Test Accuracy (R^2): {test_accuracy:.2f}")
    
    ax = plt.gca()
    ax.yaxis.set_major_formatter(FuncFormatter(format_m))

    plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
X = X.values.reshape(-1, 1)  
y = y.values.reshape(-1, 1) 

def find_best_ridge_hyperparameters(X, y, alphas, cv=5, test_size=0.2):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    ridge = Ridge()
    param_grid = {'alpha': alphas}

    grid_search = GridSearchCV(ridge, param_grid, cv=cv, scoring='r2', return_train_score=True)
    grid_search.fit(X_train, y_train)

    best_alpha = grid_search.best_params_['alpha']
    best_ridge_model = grid_search.best_estimator_

    training_predictions = best_ridge_model.predict(X_train)
    training_accuracy = r2_score(y_train, training_predictions)

    test_predictions = best_ridge_model.predict(X_test)
    test_accuracy = r2_score(y_test, test_predictions)

    return best_ridge_model, best_alpha, training_accuracy, test_accuracy

alphas = [0.1, 1, 10, 100]

best_ridge_model, best_alpha, training_accuracy, test_accuracy = find_best_ridge_hyperparameters(X, y, alphas)

print("Best Ridge Model:", best_ridge_model)
print("Best Alpha Value:", best_alpha)
print("Training Accuracy (R2):", training_accuracy)
print("Test Accuracy (R2):", test_accuracy)
```

```python
def find_best_ridge_hyperparameters(X, y, alphas, cv=5, ignore_first_x=0):
    

    # Ignore the first x data points
    X = X[ignore_first_x:]
    y = y[ignore_first_x:]

    # Define the Ridge regression model
    ridge = Ridge()

    # Define the hyperparameter grid for grid search
    param_grid = {'alpha': alphas}

    # Perform grid search with cross-validation
    grid_search = GridSearchCV(ridge, param_grid, cv=cv, scoring='r2', return_train_score=True)
    grid_search.fit(X, y)

    # Get the best hyperparameters and the corresponding model
    best_alpha = grid_search.best_params_['alpha']
    best_ridge_model = grid_search.best_estimator_

    # Calculate training accuracy (R2) for the best combination
    training_predictions = best_ridge_model.predict(X)
    training_accuracy = r2_score(y, training_predictions)

    # Calculate test accuracy (R2) for the best combination
    X_test, y_test = X[-len(y):], y
    test_predictions = best_ridge_model.predict(X_test)
    test_accuracy = r2_score(y_test, test_predictions)
    print("Best Ridge Model:", best_ridge_model)
    print("Best Alpha Value:", best_alpha)
    print("Training Accuracy (R2):", training_accuracy)
    print("Test Accuracy (R2):", test_accuracy)
    return best_ridge_model, best_alpha, training_accuracy, test_accuracy

```

```python
def format_mix(value, _):
    if value >= 1e6:
        return f'{value / 1e6:.1f}M'
    elif value >= 1e3:
        return f'{value / 1e3:.0f}K'
    else:
        return f'{value:.0f}'

def plot_actual_vs_predicted(X, y, alpha, points_to_ignore):
    if points_to_ignore > 0:
        X = X[points_to_ignore:]
        y = y[points_to_ignore:]

    ridge = Ridge(alpha=alpha)
    ridge.fit(X.reshape(-1, 1), y)

    future_years = np.arange(max(X) + 1, max(X) + 18)
    future_population = ridge.predict(future_years.reshape(-1, 1))

    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, label='Actual Population Data', color='blue')
    plt.plot(future_years, future_population, linestyle='--', label='Predicted Population Data', color='red')

    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title('Actual vs. Predicted Population Data (Ridge Regression)')
    plt.grid(True)
    plt.legend(loc='upper right', bbox_to_anchor=(1.32, 1.02))
    ax = plt.gca()
    ax.yaxis.set_major_formatter(FuncFormatter(format_mix))
    plt.show()
```
